* To use the same, we need to feed the objected detected as input to the DeepSort model.
* Here we first converted our trained model to Tensorflow framework. Used the code from below reference. The code is customised to read our trained weights and three classes based on vehicle size.

* DeepSort alogorithm uses methods available from TensorFlow version1. Code modified to adapt to use those methods without compatibility issue while running the code.

```python
# define helper function to display videos
import io 
from IPython.display import HTML
from base64 import b64encode
def show_video(file_name, width=640):
  # show resulting deepsort video
  mp4 = open(file_name,'rb').read()
  data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
  return HTML("""
  <video width="{0}" controls>
        <source src="{1}" type="video/mp4">
  </video>
  """.format(width, data_url))
```
